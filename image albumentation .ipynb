```python
def augment(aug, image):
    return aug(image=image)['image']
```

```python
lsit_aug=[ShiftScaleRotate(p=0.3, scale_limit=0.25, border_mode=1, rotate_limit=90),
                                          HorizontalFlip(p=0.8),
                                          RandomBrightnessContrast(p=0.3, brightness_limit=0.25, contrast_limit=0.5),
                                          MotionBlur(p=.2),
                                          GaussNoise(p=.2),
                                          JpegCompression(p=.2, quality_lower=50),
                                          Normalize()]

import matplotlib.image as mpimg
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]
from matplotlib.pyplot import figure
def process(filename,aug,title):
    """
    View multiple images stored in files, stacking vertically

    Arguments:
        filename: str - path to filename containing image
    """
    image = filename
    image=augment(aug,np.array(image))
    print(title)
    plt.figure(figsize=(2,2))
    plt.title(title)
    plt.imshow(image)

for i,j in enumerate(lsit_aug):
    process(image,j,str(j).split('(')[0])
```
